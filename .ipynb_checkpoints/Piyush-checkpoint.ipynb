```python
import pandas as pd
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt
import numpy as np
import face_recognition
import tensorflow as tf
from tensorflow import keras
imdir = 'Face_Mask_Detection/'
ext = ['png', 'jpg', 'gif']    # Add image formats here

files = []
[files.extend(glob.glob(imdir + '*.' + e)) for e in ext]

images = [cv2.imread(file) for file in files]
```

```python

images = np.asarray(images)
```

```python
images.shape
```

```python
dict = {"x1":[],"x2":[],"y1":[],"y2":[],"pred":[]}
# ar = []
```

```python
# This function will take image as input and detect face. 
# If face is detected then there is high probability of no mask.
# I am doing this to create label for the images.
def check(img):
    nomask = "No mask"
    mask = "mask"
    faceloc = face_recognition.face_locations(img)
    if faceloc != []:
        faceloc = faceloc[0]
        cv2.rectangle(img,(faceloc[3],faceloc[0]),(faceloc[1],faceloc[2]),(255,0,255),2)
        dict["x1"] += [faceloc[0]]
        dict["x2"] += [faceloc[1]]
        dict["y1"] += [faceloc[2]]
        dict["y2"] += [faceloc[3]]
        dict["pred"]  += [0]
        cv2.putText(img,"No Mask", (faceloc[2] , faceloc[1] ), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
    else:
        dict["x1"] += [0]
        dict["x2"] += [0]
        dict["y1"] += [0]
        dict["y2"] += [0]
        dict["pred"] += [1]
        
```

```python
#Applying function to all images.
# There will be a boundary box around face and no mask will be displayed on image.
for i in range(0,684):
    check(images[i])
```

```python
# Lets check 
plt.imshow(images[0])
```

```python
df = pd.DataFrame(dict)
```

```python
# We have the face locations as feature and the prediction as label(0 for no mask and 1 for mask)
df
```

```python
nomask = len(df[df["pred"] == 1]) # Total images which are having nomask
mask = len(df[df["pred"] == 0]) # Total images which are having mask
labels = ["mask","nomask"]
 
data = [mask,nomask]
 
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = labels)
 
# show plot
plt.show()
```

```python
#Normalization 
x1_max = df["x1"].max()
x2_max = df["x2"].max()
y1_max = df["y1"].max()
y2_max = df["y2"].max()
df["x1"] = df["x1"]/x1_max
df["x2"] = df["x2"]/x2_max
df["y1"] = df["y1"]/y1_max
df["y2"] = df["y2"]/y2_max
```

```python
df
```

```python
images[0].shape
```

```python
# I am using simple neural network with 2 layers.Input layer and output layer.
model=keras.Sequential(
    [
        keras.layers.Dense(10,input_shape=(4,),activation='relu'),
        keras.layers.Dense(1,activation='sigmoid')
    ]
)
model.compile(optimizer='adam',loss='BinaryCrossentropy',metrics=['accuracy'])
 
model.summary()
```

```python
x=df.drop('pred',axis='columns')
y=df.pred
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```
